```python
import pandas as pd
gme_revenue = pd.DataFrame({
    "Date": [
        "2024-06-30", "2024-03-31", "2023-12-31", "2023-09-30", "2023-06-30",
        "2023-03-31", "2022-12-31", "2022-09-30", "2022-06-30", "2022-03-31"
    ],
    "Revenue": [
        2110, 2100, 2125, 2075, 1980,
        2020, 1950, 1800, 1750, 1700
    ]
})
gme_revenue.to_csv("gme_revenue.csv", index=False)
gme_revenue.tail()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
def make_graph(stock_df, revenue_df, title):
    stock_df['Date'] = pd.to_datetime(stock_df['Date'])
    revenue_df['Date'] = pd.to_datetime(revenue_df['Date'])
    stock_df = stock_df.sort_values('Date')
    revenue_df = revenue_df.sort_values('Date')
    plt.figure(figsize=(8,3))
    ax1 = plt.subplot(2,1,1)
    ax1.plot(stock_df['Date'], stock_df['Close'], color='blue')
    ax1.set_title(f"{title} - Stock Price (Close)")
    ax1.set_ylabel("Close Price (USD)")
    ax1.grid(True)
    ax2 = plt.subplot(2,1,2)
    ax2.bar(revenue_df['Date'].astype(str), revenue_df['Revenue'], color='green')
    ax2.set_title(f"{title} - Quarterly Revenue")
    ax2.set_ylabel("Revenue (USD)")
    ax2.set_xlabel("Date")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
make_graph(tesla_data, tesla_revenue, "Tesla (TSLA)")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
def make_graph(stock_df, revenue_df, title):
    stock_df['Date'] = pd.to_datetime(stock_df['Date'])
    revenue_df['Date'] = pd.to_datetime(revenue_df['Date'])
    stock_df = stock_df.sort_values('Date')
    revenue_df = revenue_df.sort_values('Date')
    plt.figure(figsize=(6,3))
    ax1 = plt.subplot(2,1,1)
    ax1.plot(stock_df['Date'], stock_df['Close'], color='blue')
    ax1.set_title(f"{title} - Stock Price (Close)")
    ax1.set_ylabel("Close Price (USD)")
    ax1.grid(True)
    ax2 = plt.subplot(2,1,2)
    ax2.bar(revenue_df['Date'].astype(str), revenue_df['Revenue'], color='green')
    ax2.set_title(f"{title} - Quarterly Revenue")
    ax2.set_ylabel("Revenue (USD)")
    ax2.set_xlabel("Date")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
make_graph(gme_data, gme_revenue, "GameStop (GME)")
```
